```python
cardapio = {
    "Temaki": 15,
    "Sushi": 45.90,
    "Tempura Teishoku": 70,
    "Lamen": 65.90,
    "Ringa": 231209321093812038
}
maxi = max(cardapio, key=cardapio.get)
mini = min(cardapio, key=cardapio.get)
for prato, preco in cardapio.items():
    print(f"{prato}: R${preco}")
print(f"o prato mais caro é {maxi}")
print(f"o prato mais caro é {mini}")
```

```python
candidatos = {
    "nomes": [],
    "idades": [],
    "cidades": [],
    "vagas": []
}
sair = ""
while sair.lower().strip() != "sim":
    candidatos["nomes"].append(input("Digite seu nome: "))
    candidatos["idades"].append(int(input("Digite sua idade: ")))
    candidatos["cidades"].append(input("Digite sua cidade: "))
    candidatos["vagas"].append(input("Digite sua vaga: "))
    sair = input("Deseja sair?(sim ou não): ")
for i in range(len(candidatos["nomes"])):
    print(f"Nome: {candidatos["nomes"][i]} | Idade: {candidatos["idades"][i]} | Cidade: {candidatos["cidades"][i]} | Vaga: {candidatos["vagas"][i]}")
```

```python
while True:
    try:
        num = float(input("Digite o 1° número: "))
        num2 = float(input("Digite o 2° número: "))
        op = input("Escolha sua operação(+, -, x, /): ")
        if op not in ("+", "-", "x", "/"):
            print("Digite uma operação valida")
            continue
    except ValueError:
        print("Digite um número!")
        continue
    try:
        if op == "+":
            print(num + num2)
        elif op == "-":
            print(num - num2)
        elif op == "x":
            print(num * num2)
        elif op == "/":
            print(num / num2)
        break
    except ZeroDivisionError:
        print("Divisão inválida!")
        break
```
