```python
# df = pd.DataFrame(estoque) 
# df
# O delimitador default é a vírgula por tem que alterar o delimitador caso ele seja outro
df = pd.read_csv('vendas.csv', sep=';') # Estrutura de: nome_do_arquivo, separador/delimitador
df
# Resumindo o df basicamente carrega os dados do arquivo e depois usa essa variavel para fazer filtros e etc
# Dataframes não alteram CSV 
```

Coluna do faturamento

```python
# Se comporta igual a um dicionario
df['faturamento'] = df['quantidade_vendida'] * df['preco_unitario']
faturamento_total = df['faturamento'].sum()
faturamento_medio = df['faturamento'].mean()
# Se for display o pandas usa o numpy pra fazer o calculo ai aparece um np.float64 (numpy, tipo, binario)
print(f"Faturamento total: {faturamento_total} \nMédio: {faturamento_medio}") 
```

```python
df.to_csv('vendas_python.csv', sep=';', index=False) # Estrutura parecida com o read_csv, primeiro vem o nome depois bem o separador e por ultimo o index.
```
