```python
nomes = []
idades = []
telefones = []
emails = []
formacoes = []
for i in range(10):
    nome = input("Digite seu nome: ")
    idade = int(input("Digite sua idade: "))
    if idade < 18:
        print("Você não pode se cadstrar.")
    else:
        # Telefone e etc aqui pra ficar mais facil de testar
        telefone = input("Digite seu telefone: ")
        email = input("Digite seu email: ")
        formacao = input("Digite sua formação: ")
        nomes.append(nome)
        idades.append(idade)
        telefones.append(telefone)
        emails.append(email)
        formacoes.append(formacao)
for i in range(len(nomes)):
    print(f"Candidato {str(i+1)}: Nome: {nomes[i]} | Idade: {str(idades[i])} | Telefone = {telefones[i]} | Email: {emails[i]} | Formação: {formacoes[i]}")
```

```python
# Temperaturas classe:
temp = []
for i in range(6):
    temp.append(float(input("Digite a temperatura: ")))
for i in range(len(temp)):
    if temp[i] >= 28:
        print(f"{i+1}: A temperatura {temp[i]}° é quente.")
    elif temp[i] >= 15:
        print(f"{i+1}: A temperatura {temp[i]}° é agradável.")
    else:
        print(f"{i+1}: A temperatura {temp[i]}° é frio.")
```

```python
# Calculador de preços:
nomes = ["Nintendo Switch OLED", "Mambo", "GTA 6", "Bobas Quest", "Banana Phone"]
precos = [1499.90, 999999999, 450.50, 99.90, 1.99]
total = sum(precos)
# Depois complementa a logica ai 
maior = max(precos)
menor = min(precos)
print("Carinho: ")
for i in range(5):
    print(f"+R${precos[i]: } | {nomes[i]} foi adicionado ao carrinho")
# precos.sort() Deixei de usar o sort só pra seguir o exercicio e implementar o index
print(f"O total foi: {total} \nO item mais barato foi R${menor}: {nomes[precos.index(menor)]} \nO item mais caro foi R${maior}: {nomes[precos.index(maior)]}")
```

```python
# Adicionador de nomes
nomes = []
while True:
    nome = input("Digite um nome: ")
    nomes.append(nome)
    terminar = input("Deseja terminar?(Digite fim): ")
    if terminar.lower() == "fim":
        break
for i in range(len(nomes)):
    print(f"{i+1}: {nomes[i]}")    
```

```python
# Boletim:
nomes = ["Subaru Natsuki", "Rudeus Greyrat", "Kazuma Satou", "Izuku Midoriya", "Katsuki Bakugou", "Minoru Mineta", "Zion Vieira", "Matikanetanhauser"]
# Duas listas só pra ficar mais tranquilo
nota1 = [7.5, 8, 7.5, 9.4, 10, 2.8, 0, 10]
nota2 = [8.2, 8.4, 8.3, 9, 9.8, 1.5, 0, 10]
medias = []
aprovados = 0
recuperar = 0
reprovados = 0
for i in range(8):
    media = (nota1[i]+nota2[i])/2
    medias.append(media)
    if media >= 7:
        aprovados += 1
    elif media >= 5:
        recuperar += 1
    else:
        reprovados += 1
for i in range(8):
    print(f"{i+1} - Aluno(a) {nomes[i]} | Media: {medias[i]}")
print(f"Total de aprovados: {aprovados} | Total em recuperação: {recuperar} | Total de reprovados: {reprovados}")
```
